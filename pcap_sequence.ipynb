```python
# Used Wireshark to analyze packets
# Find first and last seq nums

from scapy.all import *
import struct

pkts = rdpcap('C:\\Users\\ritch\\Downloads\\nasdaq.pcap') 

seq_num_list = []

for packet in pkts:  # loop through each packet in pcap file
    seq_bytes = bytes(packet)[52:60]  # index position 10-18 for sequence number
    seq_num = struct.unpack(">Q", seq_bytes)[0]  # unpack binary into integers
    if seq_num < 1000000000:  # remove non-UDP packets (HSRV)
        seq_num_list.append(seq_num)

print('FIRST seq num: '+str(min(seq_num_list)))
print('LAST seq num: '+str(max(seq_num_list)))

# if you want to print all seq nums: 
# print(*seq_num_list, sep="\n")
```

```python
# Find gaps in seq nums

seq_num_list = []

for packet in pkts:  # loop through each packet in pcap file
    seq_bytes = bytes(packet)[52:60]  # index position 10-18 for sequence number
    seq_num = struct.unpack(">Q", seq_bytes)[0]  # unpack binary into integers
    if seq_num < 1000000000:  # remove non-UDP packets (HSRV)
        seq_num_list.append(seq_num)

for i in range(len(seq_num_list)-1):  # loop through the seq num list to find gaps
    j = i+1 
    if seq_num_list[j] != seq_num_list[i]+1:
        for k in range(seq_num_list[i]+1,seq_num_list[j]):
            print(k)
```
